```python
def isValid(self, s):
    stack = []
    par = {")": "(", "}": "{", "]": "["}
    for char in s:
        if char in par:
            top = stack.pop() if stack else '#'
            if par[char] != top:
                return False  
        else:
            stack.append(char)
    return not stack
```
